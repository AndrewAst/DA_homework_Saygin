```python
vnum  = vacancies.get('pages')
```

```python
all_vacancy_ids= []
for i in tqdm(range(vnum)):
    url = f'https://api.hh.ru/vacancies?text={job_title}&page={page}&per_page={num_per_page}&area={area}'
    res = requests.get(url)
    vacancies = res.json()
    vacancy_ids = [el.get('id') for el in vacancies.get('items')]
    all_vacancy_ids.extend(vacancy_ids)
```

```python
len(all_vacancy_ids)
```

```python
vacs = []
for vac_id in tqdm(all_vacancy_ids):
    url = f'https://api.hh.ru/vacancies/{vac_id}'
    res = requests.get(url)
    vacs.append(res.json())
```

```python
levels_dict = {
    'Junior': r'(junior|джун|начинающ|стаж[её]р)',
    'Middle': r'(middle|\Wмидл|миддл)',
    'Senior': r'(senior|синьор|сеньор|\Wстарш)',
    'Lead': r'(\Wlead\W|\Whead\W|\Wлид\W|\Wведущ)',
}

for grade in levels_dict.keys():
    df[grade] = df['description'].str.contains(levels_dict[grade])
    print(f"Нашлось вакансий {grade}: {df[grade].sum()}")
```

```python
jun = df['Junior'].value_counts()
mid = df['Middle'].value_counts()
sen = df['Senior'].value_counts()
lead = df['Lead'].value_counts()
labels = ('Junior', 'Middle', 'Senior', 'Lead')
sizes = [jun[1], mid[1], sen[1], lead[1]]
explode = [0.1, 0.1, 0.2, 0.2]
fig, ax = plt.subplots(figsize =(16, 10))
ax.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Соотношение по грейдам')
plt.show()
```

```python
df_sal['Junior'] = (-df_sal['Junior']).astype(np.int8)
df_sal['Middle'] = (-df_sal['Middle']).astype(np.int8)
df_sal['Senior'] = (-df_sal['Senior']).astype(np.int8)
df_sal['Lead'] = (-df_sal['Lead']).astype(np.int8)
df_j = df_sal[df_sal['Junior'] != 1]
df_m = df_sal[df_sal['Middle'] != 1]
df_s = df_sal[df_sal['Senior'] != 1]
df_l = df_sal[df_sal['Lead'] != 1]
```

```python
mean_j = df_j['salary.from'].mean()
mean_m = df_m['salary.from'].mean()
mean_s = df_s['salary.from'].mean()
mean_l = df_l['salary.from'].mean()

height = [mean_j, mean_m, mean_s, mean_l]
bars = ['Junior', 'Middle', 'Senior', 'Lead']
y_pos = np.arange(len(bars))
fig, ax = plt.subplots(figsize=(14, 10))
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.title('Средняя зарплата в зависимости от грейда')
plt.show()
```
