```python
vnum  = vacancies.get('pages')
```

```python
all_vacancy_ids= []
for i in tqdm(range(vnum)):
    url = f'https://api.hh.ru/vacancies?text={job_title}&page={page}&per_page={num_per_page}&area={area}'
    res = requests.get(url)
    vacancies = res.json()
    vacancy_ids = [el.get('id') for el in vacancies.get('items')]
    all_vacancy_ids.extend(vacancy_ids)
```

```python
len(all_vacancy_ids)
```

```python
vacs = []
for vac_id in tqdm(all_vacancy_ids):
    url = f'https://api.hh.ru/vacancies/{vac_id}'
    res = requests.get(url)
    vacs.append(res.json())
```

```python
labels = [x for x in df['employment.name'].unique()]
sizes = df['employment.name'].value_counts().reset_index()['employment.name']
explode = [0.4, 0.4, 0.6, 0.6]
fig, ax = plt.subplots(figsize =(20, 14))
ax.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Распределение вакансий по типу занятости')
plt.show()
```

По полученным данным 96% представлены постоянной занятостью, компаниям необходим штатный сотрудник, скажем так находящийся всегда под рукой. 1% вовлечены в проектную работу, что говорит о том, что инновационные работы в РФ на рынке Data ведутся, но проект - всегда риск, потому не готовы гарантировать постоянную занятость. Очень расстраивает показатель стажировки, лишь 1% работодателей готовы предоставить стажировку, что говорит об экономии ресурсов в компаниях, никто не хочет тратить ни время, ни деньги на воспитание новоиспеченных специалистов.

```python
# Распределение вакансий по городам
labels = ['Москва', 'Санкт-Петербург', 'Казань', 'Екатеринбург', 'Новосибирск', 'Пермь', 'Краснодар', 'Ростов-на-Дону', 'Оренбург', 'Владимир', 'Уфа', 'Псков', 'Луховицы', 'Киров (Кировская область)', 'Саратов', 'Самара', 'Ижевск', 'Пенза', 'Воронеж', 'Брянск', 'Челябинск']
sizes = df['area.name'].value_counts().reset_index()['area.name']
fig, ax = plt.subplots(figsize =(20, 14))
ax.pie(sizes, labels=labels, autopct='%1.1f%%',startangle=190)
ax.axis('equal')
ax.legend(labels, loc='lower left')
plt.title('Распределение вакансий по городам')
plt.show()
```

Как мы видим из представленного графика наибольшее количество вакансий представлено в г.Москва, что логично - головные компании располагаются в столице, дата-центры и т.д. стараются держать в непосредственной близости. На 2 месте - Санкт-Петербург, полагаю связано с тем, что в Москве ценник на услуги IT специалистов ниже чем в Москве, плюс расположен в непосредственной близости от Москвы.

```python
df_skil = df[df['key_skills'] != 0]
```

```python
skills = df_skil['key_skills']
```

```python
q=[]
for i in skills.keys():
  if len(skills[i])> 0: 
    q.append(skills[i])
    print(len(skills[i]))
```

```python
i=0
j=0
w=[]
```

```python
while i<len(q):
  while j<len(q[i]):
    w.append(q[i][j].get('name'))
  #  print(q[i][j])
    j+=1
  j=0
  i+=1
```

```python
len(w)
```

```python
z=pd.DataFrame({'Skill':[], 'count':[]})
```

```python
i=0 
while i<len(w):
  new_row={'Skill':w[i],'count':w.count(w[i])}
  z=z.append(new_row, ignore_index=True)
 # print(w[i]+''+str(w.count(w[i])))
  i+=1
```

```python
zz=z.drop_duplicates().sort_values('count',ascending=False)
```

```python
fig = px.pie(zz, values='count', names='Skill',
             color ='Skill', color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.update_traces(textposition='inside',
                  textinfo='percent+label',
                  sort=False)
fig.update_layout(width=1800, height=900)
fig.show()
```
